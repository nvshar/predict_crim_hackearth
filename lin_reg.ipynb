```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("criminal_train.csv")
X=df.drop(columns=['PERID','Criminal'])
y=df['Criminal']
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(X,y)
```

```python
df_t=pd.read_csv('criminal_test.csv')
df_t1=df_t.drop(columns='100100')
pre=lm.predict(df_t1)
maxx=np.max(pre)
pre1=pre/maxx
pre2=np.round(pre1)
```

```python
for i in range(11430):
    if pre2[i]==-0:
        pre2[i]=0
```

```python
df_t = df_t.rename(columns={'100100': 'PERID'})
```

```python
dataframe1=df_t
dataframe1['Criminal']=pre2
db=dataframe1.iloc[:,[0,71]]
#pd.read_csv('prediction.csv')
```
